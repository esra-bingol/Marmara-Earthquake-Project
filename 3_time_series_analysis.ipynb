## Hedefler:
- YÄ±llara gÃ¶re toplam deprem sayÄ±sÄ±
- Ortalama bÃ¼yÃ¼klÃ¼k (ML) deÄŸiÅŸimi
- Opsiyonel: AylÄ±k sezonsal trend


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("marmara_faults_earthquakes_2000_2025.csv")
df['Date'] = pd.to_datetime(df['Date'])

# Marmara filtreleme
df_marmara = df[(df['Latitude'] > 39) & (df['Latitude'] < 42) & 
                (df['Longitude'] > 26) & (df['Longitude'] < 31)].copy()
```

```python
# YÄ±l sÃ¼tunu oluÅŸumu
df_marmara['Year'] = df_marmara['Date'].dt.year

# YÄ±llÄ±k toplam deprem sayÄ±sÄ±
yearly_counts = df_marmara.groupby('Year').size()

# GÃ¶rselleÅŸtirme adÄ±mÄ±
plt.figure(figsize=(10, 5))
sns.lineplot(x=yearly_counts.index, y=yearly_counts.values, marker='o')
plt.title('Marmara BÃ¶lgesi - YÄ±llÄ±k Deprem SayÄ±sÄ± (2000â€“2025)')
plt.xlabel('YÄ±l')
plt.ylabel('Deprem SayÄ±sÄ±')
plt.grid(True)
plt.show()
```
